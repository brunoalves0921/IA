```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leitura do arquivo CSV
df = pd.read_csv('bloco_1.csv')

# Obtenção das taxas de elitismo únicas
taxas_elitismo = df['taxaElitismo'].unique()

# Configuração do gráfico
plt.figure(figsize=(10, 6))
plt.title('Melhor Fitness ao longo das Iterações por Taxa de Elitismo')
plt.xlabel('Iteração')
plt.ylabel('Melhor Fitness')

# Plotagem das linhas para cada taxa de elitismo
for taxa in taxas_elitismo:
    dados_taxa = df[df['taxaElitismo'] == taxa]
    plt.plot(dados_taxa['iteracao'], dados_taxa['melhorFitness'], label=f'Taxa de Elitismo {taxa}')

# Adição de legenda e exibição do gráfico
plt.gca().invert_yaxis()
plt.legend()
plt.grid(True)
plt.show()
```
