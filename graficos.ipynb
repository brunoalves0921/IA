```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leitura do arquivo CSV
df = pd.read_csv('bloco_1.csv')

# Obteno das taxas de elitismo nicas
taxas_elitismo = df['taxaElitismo'].unique()

# Configurao do grfico
plt.figure(figsize=(10, 6))
plt.title('Melhor Fitness ao longo das Iteraes por Taxa de Elitismo')
plt.xlabel('Iterao')
plt.ylabel('Melhor Fitness')

# Plotagem das linhas para cada taxa de elitismo
for taxa in taxas_elitismo:
    dados_taxa = df[df['taxaElitismo'] == taxa]
    plt.plot(dados_taxa['iteracao'], dados_taxa['melhorFitness'], label=f'Taxa de Elitismo {taxa}')

# Adio de legenda e exibio do grfico
plt.gca().invert_yaxis()
plt.legend()
plt.grid(True)
plt.show()
```
